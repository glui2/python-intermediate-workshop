# GroupBy
If you're using for loops with pandas, you're doing it wrong

```python
cars = pd.read_csv("D:Data/Auto.csv")
cars
```

```python
cars.groupby("origin")['weight'].mean() # split by origin, calculated mean of the respective weights columns, then combine
```

```python
cars['weight'].mean() # mean of specific column
```

```python
cars.mean() # mean of every column
```

```python
cars
```

```python
cars.groupby('origin').mean() # categorise before applying groupby, split by 'origin', apply mean of every column, then combine
```

```python
cars['heavy'] = cars['weight'] > cars['weight'].mean()
cars.groupby('heavy').mean()
```

```python
counts = cars.groupby('cylinders')['origin'].value_counts().unstack() # split based on cylinders, then apply value_counts().unstack() on 'origin', then combine 
```

## Page 100 Exercises
rainfall -> f() -> most recent date that this much rain occurred

```python
# question 1
def last_occurrence_of_rainfall(rainfall):
    rain = pd.read_csv(filename, names=["Date","Rainfall"], header=None, parse_dates=["Date"], dayfirst=True)
    rain.set_index("Date", inplace=True)
    rainfall_mask = rain['Rainfall'] >= rainfall
    return rain.loc[rainfall_mask].index.max()
     
```

```python
last_occurrence_of_rainfall(500)
```
